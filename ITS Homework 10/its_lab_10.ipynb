```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

Load the training xml file provided with the material of this lab using the function

```
faceCascade = cv2.CascadeClassifier('haarcascade_frontalface.xml')
```

Using mpimg instead of imread to load the image, which will handle numpy image manipulations, so i dont have to grayscale the image or use rgb conversions when viewing or drawing rectangles

```
image = mpimg.imread('ubuntu_team.jpg')

plt.figure(figsize=(8,6))
plt.imshow(image)
```

Detects the faces in the picture

```
faces = faceCascade.detectMultiScale(image, scaleFactor=1.1, minNeighbors=1)
```

```
#number of faces detected
len(faces)
```

Find all the faces in the picture

```
#loop over to extract detected faces
for (x, y, w, h) in faces:
  
  #draw a rectangle
  cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 1)

plt.figure(figsize=(8,6))
plt.imshow(image)
```

Detect the face using the webcam in the laptop.

```
#video
cap = cv2.VideoCapture(0)

while(True):
    ret, image_frame = cap.read()
    
    faces = faceCascade.detectMultiScale(image_frame, scaleFactor=1.1, minNeighbors=1)
        
    #loop over to extract detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(image_frame, (x, y), (x + w, y + h), (255, 0, 0), 1)
        
    cv2.imshow('frame', image_frame)
    if cv2.waitKey(1) == ord('q'):
      break

#when everything done, release the capture
cap.release()
cv2.destroyAllWindows()

#you will get a traceback error when trying to quit the program, it can be handle but I left it as it was not that important for the homework.
#Anyway you will not get that error if you run this as a .py file in a python IDE
```
