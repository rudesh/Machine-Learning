## **Question 03**



*No pictures are not identical*

```python
# reading Lena-A and Lena-B images from file

LenaA = cv2.imread('Lena-A.jpg')
LenaB = cv2.imread('Lena-B.jpg')
```

```python
# checking if they are identical by using below method
# and in the result we can clearly see that they are not identical, shape is the property that shows the size of matrix and bitwise_xor is as the name suggests

if LenaA.shape == LenaA.shape and not(np.bitwise_xor(LenaA,LenaB).any()):
    print('Identical')
else:
    print('Not Identical')
```

## **Question 04**

```python
#convert LenaA and LenaB to grayscale

imageAG = cv2.cvtColor(LenaA, cv2.COLOR_RGB2GRAY)
imageBG = cv2.cvtColor(LenaB, cv2.COLOR_RGB2GRAY)
```

```python
# for this exercise i have used the sample code provided and modified it to identify green and red coloured dots


image = cv2.imread('YywIS.jpg')

# define range of blue color in HSV
lower_blue = np.array([110,50,50])
upper_blue = np.array([130,255,255])

lower_green = np.array([65,50,50])
upper_green = np.array([80,255,255])

lower_red = np.array([125,100,100])
upper_red = np.array([220,255,255])


# do the color conversion
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Threshold the HSV image to get only blue colors
mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
mask_green = cv2.inRange(hsv, lower_green, upper_green)
mask_red = cv2.inRange(hsv, lower_red, upper_red)

# since its not neccesary, I have not used the save the images calls
#cv2.imwrite('/content/gdrive/My Drive/Colab Notebooks/ITS Labs/Lab 02/HSV.jpg', hsv)
#cv2.imwrite('/content/gdrive/My Drive/Colab Notebooks/ITS Labs/Lab 02/blue_mask.jpg', mask)
```
